```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Language Tooling
from nltk import download
from nltk.corpus import stopwords
from wordcloud import WordCloud

# Machine Learning
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
```

```python
data_sentimen = pd.read_csv(
    'https://raw.githubusercontent.com/radya13rahman/Sentiment_Analysis_Sekolah_Daring/master/Dataset/datatest.csv', 
    sep=';')

data_sentimen.head()
```

```python
data_sentimen.info()
```

```python
# Stopwords init
download('stopwords')
stop = stopwords.words('indonesian')
```

```python
data_sentimen['tweet']
```

```python
stop.extend(['yg', 'aja', 'sih', 'sekolah', 'banget', 'online'])
```

```python
# Mengubah Huruf Besar Menjadi Huruf Kecil
data_sentimen['tweet'] = data_sentimen['tweet'].str.lower()
```

```python
# Hilangkan stopwords
data_sentimen['tweet'] = data_sentimen['tweet'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

###Model Building

```python
X_train, X_test, y_train, y_test = train_test_split(data_sentimen['tweet'], data_sentimen['sentiment'], test_size=0.3)
```

```python
X_train
```

```python
cvec = CountVectorizer()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
model = Pipeline(
    [('vectorizer', cvec), 
     ('classifier', knn)])

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plt.bar(10, len(np.where(y_pred == 'positive')[0]), label='positive')
plt.bar(15, len(np.where(y_pred == 'negative')[0]), label='negative')
plt.bar(20, len(np.where(y_pred == 'neutral')[0]), label='neutral')

plt.legend(loc='upper right')
plt.xlabel('Kategori Sentiment')
plt.ylabel('Jumlah Data')

plt.show()
```

```python
# Membuat model tanpa sentimen netral
new_data = data_sentimen[data_sentimen['sentiment'] != 'neutral'].copy()
```

```python
x_train, x_test, Y_train, Y_test = train_test_split(new_data['tweet'], new_data['sentiment'], test_size = 0.3, random_state=22)
```

```python
models = Pipeline(
    [('vectorizer', cvec), 
     ('classifier', knn)])

models.fit(x_train, Y_train)
```

```python
Y_pred = models.predict(x_test)
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

```python
for twt in data_sentimen[data_sentimen['sentiment'] == 'neutral']['tweet']:
  new_sentimen = models.predict([twt])[0]
  new_data = new_data.append({'tweet': twt, 'sentiment': new_sentimen}, ignore_index = True)
```

```python
train_X, test_X, train_y, test_y = train_test_split(new_data['tweet'], new_data['sentiment'], test_size = 0.3, random_state=22)
```

```python
sentimen_model = Pipeline(
    [('vectorizer', cvec), 
     ('classifier', knn)])

sentimen_model.fit(train_X, train_y)
```

```python
pred = sentimen_model.predict(test_X)
pred
```

```python
print(confusion_matrix(test_y, pred))
print(classification_report(test_y, pred))
```
